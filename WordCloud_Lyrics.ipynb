# WordCloud Project 

 This project is to analyze the lyrics of the rapper Nicki Minaj, to analyze what are th emost common words used in her music. The lyrics were aggregarted from major record releases, ignoring any collaborative songs on different artists albums. The lyrics are from albums up to 2018. The aggregated lyrics are in file 'minaj-lyrics.txt'. 
 This project was originally made on July 19 2020.

## Import libraries

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# to install wordcloud       
# !pip install wordcloud 
```

## Read in the text file

```python
# read the whole text
text = open('minaj-lyrics.txt').read() # 

# generate a wordcloud and plot it
wordcloud = WordCloud(width=2000, 
                      height=2000, 
                      background_color='black', 
                      colormap='Blues', 
                      repeat=False, 
                      max_font_size=400, 
                      min_font_size=40).generate(text)

plt.imshow(wordcloud)

# turn off axis - avoid numbers around cloud
plt.axis('off')

# show
plt.show()
```
